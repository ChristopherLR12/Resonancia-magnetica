```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation
```

```python
#Constantes
m0 = 1
B0 = 3
B1 = 2e-3
γ = (42.576e6)#/np.pi
w0 = γ*B0
w1 = γ*B1

def M(t):
    return (m0*np.sin(w1*t)*np.sin(w0*t),
            m0*np.sin(w1*t)*np.cos(w0*t),
            m0*np.cos(w1*t))

tt = np.linspace(0,0.001,5000)

xx = [M(t)[0] for t in tt]
yy = [M(t)[1] for t in tt]
zz = [M(t)[2] for t in tt]


#Datos
ds=np.array([xx,yy,zz])

#Figura
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

# Definición de los límites de los ejes

x_min, x_max = ds[0].min(), ds[0].max()
y_min, y_max = ds[1].min(), ds[1].max()
z_min, z_max = ds[2].min(), ds[2].max()

def animate_func(num):
    
    ax.clear()
    
    for i in range(num-1):
        
        # Se añade la ubicación del punto en el paso num. Se actualiza en cada iteración.
        ax.scatter(ds[0, i], # eje x
                   ds[1, i], # eje y
                   ds[2, i], # eje z
                   c='k', # Color del punto
                   marker='o',
                   s=20)
    
    ax.set_title("Evolución de magnetización",size=20,fontweight="bold")
    
    ax.set_xlabel('Eje x')
    ax.set_ylabel('Eje y')
    ax.set_zlabel('Eje z')
    
    ax.set_xlim3d([x_min, x_max])
    ax.set_ylim3d([y_min, y_max])
    ax.set_zlim3d([z_min, z_max])

    plt.show()


anim = animation.FuncAnimation(fig, # Figura donde se plotea la animación.
                               animate_func, #  Función de actualización de la animación, itera sobre num.
                               interval=100, #Delay ms
                               frames=185 # Número de frames a mostrar.
                              )

anim.save('anim3.gif', writer="Pillow")
plt.show()
```
